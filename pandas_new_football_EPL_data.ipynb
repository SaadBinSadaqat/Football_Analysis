```python
# Calculate the total number of goals scored with left and right foot
left_foot_goals = df['Goals with left foot'].sum()
right_foot_goals = df['Goals with right foot'].sum()

# Create a bar chart to visualize the distribution of goals scored by players' dominant foot
plt.figure(figsize=(8, 6))
bars = plt.bar(['Left Foot', 'Right Foot'], [left_foot_goals, right_foot_goals], color=['skyblue', 'salmon'])
plt.title('Distribution of Goals by Dominant Foot')
plt.xlabel('Dominant Foot')
plt.ylabel('Number of Goals')
plt.xticks(rotation=45)

# Add labels on each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval), va='bottom')

plt.tight_layout()
plt.show()

print("Left Foot Goals:", left_foot_goals)
print("Right Foot Goals:", right_foot_goals)
```

```python
# Filter out players who have scored penalties
penalty_scorers = df[df['Penalties scored'] > 0]

# Sort the penalty scorers by the number of penalties scored in descending order
top_penalty_scorers = penalty_scorers.sort_values(by='Penalties scored', ascending=False).head(10)

# Create a bar plot to visualize the top penalty scorers
plt.figure(figsize=(10, 6))
sns.barplot(x='Penalties scored', y='Name', data=top_penalty_scorers, hue='Name', palette='viridis', legend=False)
plt.title('Top Penalty Scorers')
plt.xlabel('Number of Penalties Scored')
plt.ylabel('Player Name')
plt.tight_layout()
plt.show()
```

```python
cards_df = df[['Name', 'Red cards', 'Yellow cards']]
cards_df = cards_df.groupby('Name').sum().reset_index()

# Sort the DataFrame by the total number of cards received in descending order
cards_df['Total cards'] = cards_df['Red cards'] + cards_df['Yellow cards']
cards_df = cards_df.sort_values(by='Total cards', ascending=False).head(10)

# Melt the DataFrame to combine red and yellow cards into a single column
cards_df_melted = pd.melt(cards_df, id_vars='Name', value_vars=['Red cards', 'Yellow cards'], var_name='Card Type', value_name='Number of Cards')

# Create a bar plot to visualize the players with the most red and yellow cards
plt.figure(figsize=(10, 6))
sns.barplot(y='Number of Cards', x='Name', hue='Card Type', data=cards_df_melted, palette='muted')
plt.title('Players with Most Red and Yellow Cards')
plt.ylabel('Number of Cards')
plt.xlabel('Player Name')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset - 2020-09-24.csv')

country_counts = df['Nationality'].value_counts()

# Select the top 5 countries with the most representation
top_5_countries = country_counts.head(5)

# Plot the data using a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_countries, labels=top_5_countries.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Countries with Most Representation in the League')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import plotly.express as px

position_counts = df.groupby(['Nationality', 'Position']).size().reset_index(name='Count')
position_counts_pivot = position_counts.pivot(index='Nationality', columns='Position', values='Count').fillna(0)
position_counts_pivot['Total'] = position_counts_pivot.sum(axis=1)

sunburst_data = pd.DataFrame({
    'Position': ['Total'] * len(position_counts_pivot.index),
    'Nationality': position_counts_pivot.index,
    'Count': position_counts_pivot['Total']
})

# Ensure 'sunburst_data' has the same structure as 'df'
sunburst_data = sunburst_data.reindex(columns=df.columns)

concatenated_df = pd.concat([df, sunburst_data])

# Filter out non-leaf rows
leaf_rows = concatenated_df[concatenated_df['Nationality'].notna()]

fig = px.sunburst(leaf_rows,
                  path=['Position', 'Nationality'],
                  values='Appearances'
                 )

fig.update_layout(
    title='<b>Players position by country</b>',
    title_x=0.5,
    titlefont=dict(color='black', size=28, family="Courier New, monospace"),
    width=600,
    height=600,
    showlegend=False
)

fig.show()
```
